```python
import numpy as np
import pandas as pd
```

```python
file = pd.read_csv('filename.txt',header=None)
```

```python
file_name = file.values
```

```python
import pandas as pd
import sys
from scipy.stats import ks_2samp

#first_file = sys.argv[1]
#second_file = sys.argv[2]

def corr(first_file, second_file):
    # assuming first column is `class_name_id`
    first_df = pd.read_csv(first_file, index_col=0)
    second_df = pd.read_csv(second_file, index_col=0)
    class_names = ['toxic', 'severe_toxic', 'obscene', 'threat', 'insult', 'identity_hate']

    ## toxic
    v1 = first_df["toxic"].corr(second_df["toxic"], method='pearson')
    print(first_file[:10],second_file[:10])
    print(' Pearson\'s correlation score: %0.6f' % v1)
    v2 = first_df["severe_toxic"].corr(second_df["severe_toxic"], method='pearson')
    v3 = first_df["obscene"].corr(second_df["obscene"], method='pearson')
    v4 = first_df["threat"].corr(second_df["threat"], method='pearson')
    v5 = first_df["insult"].corr(second_df["insult"], method='pearson')
    v6 = first_df["identity_hate"].corr(second_df["identity_hate"], method='pearson')
    return(v1,v2,v3,v4,v5,v6)
```

```python
r1=np.zeros((len(file_name),len(file_name)))
r2=np.zeros((len(file_name),len(file_name)))
r3=np.zeros((len(file_name),len(file_name)))
r4=np.zeros((len(file_name),len(file_name)))
r5=np.zeros((len(file_name),len(file_name)))
r6=np.zeros((len(file_name),len(file_name)))

for i in range(len(file_name)):
    for j in range(len(file_name)):
        v1,v2,v3,v4,v5,v6=corr(file_name[i][0],file_name[j][0])
        r1[i,j]=v1
        r2[i,j]=v2
        r3[i,j]=v3
        r4[i,j]=v4
        r5[i,j]=v5
        r6[i,j]=v6
```

```python
s1= pd.DataFrame(r1)
s2= pd.DataFrame(r2)
s3= pd.DataFrame(r3)
s4= pd.DataFrame(r4)
s5= pd.DataFrame(r5)
s6= pd.DataFrame(r6)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
fig=plt.figure(figsize=(22,22))
ax1 = fig.add_subplot(321)
sns.heatmap(s1, annot=True, cmap="YlGnBu",linewidths='0.5',ax=ax1)
plt.title("Toxic",fontsize=20)
#plt.xlabel("Date",fontsize=18)
#plt.ylabel("Month",fontsize=18)

ax2 = fig.add_subplot(322)
sns.heatmap(s2,annot=True, cmap="YlGnBu",linewidths='0.5',ax=ax2)
plt.title('severe_toxic',fontsize=20)
fig.subplots_adjust(wspace=0.001)

ax3 = fig.add_subplot(323)
sns.heatmap(s3, annot=True, cmap="YlGnBu",linewidths='0.5',ax=ax3)
plt.title("obscene",fontsize=20)
#plt.xlabel("Date",fontsize=18)
#plt.ylabel("Month",fontsize=18)

ax4 = fig.add_subplot(324)
sns.heatmap(s4,annot=True, cmap="YlGnBu",linewidths='0.5',ax=ax4)
plt.title('threat',fontsize=20)
fig.subplots_adjust(wspace=0.001)

ax5 = fig.add_subplot(325)
sns.heatmap(s3, annot=True, cmap="YlGnBu",linewidths='0.5',ax=ax5)
plt.title('insult',fontsize=20)
#plt.xlabel("Date",fontsize=18)
#plt.ylabel("Month",fontsize=18)

ax6 = fig.add_subplot(326)
sns.heatmap(s4,annot=True, cmap="YlGnBu",linewidths='0.5',ax=ax6)
plt.title('identity_hate',fontsize=20)
fig.subplots_adjust(wspace=0.001)
```
